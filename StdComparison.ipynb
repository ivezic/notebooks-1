Just plot all the metrics, using lines .. see what happens.

```python
plt.figure(figsize=(20,10))
x = np.arange(len(comparison['minion_1016']))
colors = np.random.random_sample((len(runs), 3))
for i, r in enumerate(runs):
    plt.plot(x, comparison[r]/comparison['minion_1016'], color=colors[i], label=r)
plt.xticks(x, comparison.Summary_Name, rotation=90)
plt.axhline(1.0)
#plt.ylim(0.9, 1.1)
plt.legend(loc=(1.0, 0.2))
```

Compare a more limited set of metrics. Easier to plot and read the plot.

```python
plotmetrics = ['NVisits All Visits', 'Fraction of total NVisits WFD', 
               '(days) Total effective time of survey All Visits', 
              'Median OpenShutterFraction Per night OneDSlicer']
plotcomp = comparison.query('Summary_Name in @plotmetrics')
baseline = plotcomp['minion_1016']
plotcomp
```

```python
xoffset = 0.9/(float(len(runs)))
x = np.arange(len(baseline))
colors = np.random.random_sample((len(runs), 3))
for i, r in enumerate(runs):
    plt.bar(x+i*xoffset, plotcomp[r]/baseline, width=xoffset, color=colors[i], label=r)
plt.xticks(x, plotcomp.Summary_Name, rotation=60)
for xi in x:
    plt.axvline(xi-0.05, linestyle=':')
plt.axhline(1.0)
plt.ylim(0.8, 1.2)
plt.legend(loc=(1.1, 0))
```

```python
x = np.arange(len(baseline))
colors = np.random.random_sample((len(runs), 3))
for i, r in enumerate(runs):
    plt.plot(x, plotcomp[r]/baseline, color=colors[i], label=r)
plt.xticks(x, plotcomp.Summary_Name, rotation=90)
plt.axhline(1.0)
#plt.ylim(0.9, 1.1)
plt.legend(loc=(1.1, .0))
```

Try the same metrics, but swap the axes. RunNames on x axis, metric names in legend.

```python
metricname = 'Summary_Name'

plt.figure()
x = np.arange(len(runs))
for i, m in enumerate(plotmetrics):
    tmp = plotcomp.query('Summary_Name == @m').drop('Summary_Name', axis=1).as_matrix()[0]
    b = plotcomp.query('Summary_Name == @m')['minion_1016'].iloc[0]
    plt.plot(x, tmp/b, marker='.', label=m)
plt.xticks(x, runs, rotation=90)
plt.legend(loc=(1.1, 0.2))
#plt.ylim(0.9, 1.1)
```
